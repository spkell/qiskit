```python
from qiskit import QuantumCircuit, Aer, execute
from qiskit.visualization import plot_histogram
```

```python
circ = QuantumCircuit(2,2) #2 qubits, 2 classical bits
circ.initialize([1,0],0) #initializes 1st qubit to the 0 state
circ.initialize([0,1],1) #initializes 2nd qubit to the 1 state

circ.h(0)
circ.cx(0,1)
circ.z(0)
circ.x(0) #switch 0 qubit so CNOT gate is activated when the 0 qubit is in the |0> state before this gate
circ.cx(0,1)
circ.x(0) #switch back qubit 0 to its correct state

circ.measure(0,1)
circ.measure(1,0)

circ.draw('mpl')
```

```python
sim = Aer.get_backend('qasm_simulator')
res = execute(circ, sim, shots = 1000).result()
plot_histogram(res.get_counts())
```
