```python
from qiskit import QuantumCircuit, Aer, execute
from qiskit.visualization import plot_histogram
```

```python
f = 3

circ = QuantumCircuit(3,2) #1 qubit circuit
circ.initialize([1,0],0) #initializes 1st qubit to the 0 state
circ.initialize([1,0],1) #initializes 2nd qubit to the 0 state
circ.initialize([0,1],2) #initializes 3rd qubit to the 1 state

circ.h(0)
circ.h(1)
circ.h(2)
circ.barrier()
################

if f == 0: #f0
    pass
elif f == 3:
    circ.cx(0,2)
elif f == 5:
    circ.cx(1,2)
elif f == 6:
    circ.x(0)
    circ.ccx(0,1,2)
    circ.x(0)
    circ.x(1)
    circ.ccx(0,1,2)
    circ.x(1)
elif f == 9:
    circ.x(0)
    circ.x(1)
    circ.ccx(0,1,2)
    circ.x(0)
    circ.x(1)
    circ.ccx(0,1,2)
elif f == 10:
    circ.x(1)
    circ.cx(1,2)
    circ.x(1)
elif f == 12:
    circ.x(0)
    circ.cx(0,2)
    circ.x(0)
elif f == 15:
    circ.x(2)
    
################
circ.barrier()

circ.h(0)
circ.h(1)

circ.measure(0,1)
circ.measure(1,0)

circ.draw('mpl')
```

```python
sim = Aer.get_backend('qasm_simulator')
res = execute(circ, sim, shots = 1000).result()
plot_histogram(res.get_counts())
```
